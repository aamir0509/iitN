**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```
f = open('abc.txt','r')
text = f.read()
words = text.split()
l1=[]
for i in words:  
  if (words.count(i),i) not in l1:    
    l1.append((words.count(i),i))
l1.sort(reverse=True)
#print(l1)
d=dict([i for i in l1])
print(d)
for i in range(100):
  print(l1[i])
#sorted(words, key=max, reverse=True)
```
