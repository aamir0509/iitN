**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```
f=open('abc.txt')
txt=f.read()
words=txt.split()
freq_chart={}
for word in words:
  curr_count=freq_chart.setdefault(word,1)
  freq_chart[word]=curr_count+1
freq_chart()

```
