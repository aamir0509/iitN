**Lab 2**

* Write a program that asks user to enter ten words, one at a time and then it prints those words in sorted order
* Write a program that prints the sum of all even numbers between 1 and 1000

```
n=int(input('Enter the maximum number'))
i=1
sum=0
while i<n:
  if i%2==0:
    sum=sum+i
  i=i+1
print('sum of all even numbers',sum)
```

**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```

f=open('abc.txt','r')
txt=f.read()
words=txt.split()
freq_chart={}
for word in words:
  curr_count=freq_chart.setdefault(word,0)
  freq_chart[word]=curr_count+1
sorted(freq_chart.items(),key=lambda x:x[1],reverse=True)
  
```
