**Lab 4**

* Implement accelerate and deaccelerate methods for the car class assume of fix rate of acceleration and deaccelerate make sure car doesn’t move backward when it deaccelerating.
* Simulate a race of 10 cars. All car starts at the same position and accelerates at a random rate between 1 and 5 m/$s^2$ for a period of between 5 to 10 seconds.


```
class Car_race:
  def __init__(self, speed, rate, time):
    self.speed = speed
    self.rate = rate
    self.time = time

  def start(self):
    self.state = 'started'

  def accelerate(self):
     if self.state in ['started']:
        self.speed = self.speed + rate * time
        self.position = self.position + (rate * time * time/2.0)
     else:
        pass

  def deaccelerate(self):
    self.speed = self.speed - rate * time
    self.position = self.position + (self.position - (rate * time * time/2.0))

  c = Car_race(100, 5, 5)
  c.accelerate()
  c.deaccelerate()

  



  
```
