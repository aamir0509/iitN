**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```
#import nltk
#from nltk.corpus import stopwords
#set(stopwords.words('english'))
f = open('abc.txt')
text = f.read()
words = text.split()
words = [x.lower() for x in words]
#print (words)
#words = words.lower()
freq_chart={}
for word in words:
  curr_count=freq_chart.setdefault(word,0)
  freq_chart[word]= curr_count + 1
val=sorted(freq_chart.items(),key=lambda x:x[1],reverse=True)
#print (freq_chart)
print(val[:100])

```
