**Lab 4**

* Implement accelerate and deaccelerate methods for the car class assume of fix rate of acceleration and deaccelerate make sure car doesn’t move backward when it deaccelerating.
* Simulate a race of 10 cars. All car starts at the same position and accelerates at a random rate between 1 and 5 m/$s^2$ for a period of between 5 to 10 seconds.


```
import random
class car:
  def __init__(self,name,pos=0, speed=0):
    self.name =name
    self.pos=pos
    self.speed=speed
  def accelarate(self,a,t):
    self.pos = self.pos+self.speed *t+0.5*a*t*t
    self.speed=self.speed+a*t
  def decelarate(self,d,t):
    tzero=self.speed/d
    if t<tzero:
      self.pos=self.pos+self.speed*t-.5*d*t*t
    else:
      self.pos=self.pos+self.speed*tzero-(.5*d*tzero*tzero)

c1 = car('ramesh')
c2 = car('manoj')
c3=car('rohit')
c4 = car('raghu')
car_list=[c1,c2,c3,c4]
for car in car_list:
  car.accelarate(random.randint(1,4),random.randint(5,10))
for car in car_list:
  print("%s:%s",(car.name,car.pos))


```
