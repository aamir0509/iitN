* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```


f=open('abc.txt','r')
text=f.read()
words=text.split()
freq_chart={}
for word in words:
  curr=freq_chart.setdefault(word,0)
  freq_chart[word]=curr+1
sorted(freq_chart.items(),key = lambda x:x[1],reverse=False)

```
