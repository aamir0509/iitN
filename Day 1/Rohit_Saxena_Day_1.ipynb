**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```
f = open('abc.txt', 'r')
text = f.read()
words =  text.split()
l1=[]
for i in words:
  if(words.count(i),i) not in l1:
    l1.append((words.count(i),i))
l1.sort(reverse=True)
for i in range(100):
  print(l1[i])
```

```
import random
class Car:
  def __init__(self,make,model,year,color):
    self.make = make
    self.model = model
    self.year= year
    self.color = color
    self.speed = 0
    self.position=0
    self.fuelLevel = 100
    self.state = "started"

  def start(self):
    self.state= "started"

  def accelerate(self, rate,time):
    if self.state in ["started", "running"]:
      self.speed=self.speed + rate * time
      self.position = self.position + (rate*time*time/2.0)
      self.state = "running"
    else:
      pass

  def deaccelerate(self, rate,time):
    self.speed=self.speed - rate * time
    if self.speed<0:
      self.speed=0  
    if self.speed *time > (rate*time*time/2.0):
      self.position = self.position + (self.speed *time - (rate*time*time/2.0))
    else: # to ensure that car doesn’t move backward
      pass
    
  def getcurrentposition(self): # getter function
    print("current position is: ",self.position)

  def getcurrentspeed(self): # getter function
    print("current speed is: ",self.speed)

c1=Car("Maruti", "Zen", 2004, "Blue")
c1.start()
c1.accelerate(random.randrange(1,6),random.randrange(5,11))
c2=Car("Maruti", "Accent", 2018, "White")
c2.start()
c2.accelerate(random.randrange(1,6),random.randrange(5,11))
c3=Car("Maruti", "Alto", 2002, "Gray")
c3.start()
c3.accelerate(random.randrange(1,6),random.randrange(5,11))
c4=Car("Audi", "Zeal", 2010, "Blue")
c4.start()
c4.accelerate(random.randrange(1,6),random.randrange(5,11))
c5=Car("Wolkswegan", "Zenith", 2013, "Green")
c5.start()
c5.accelerate(random.randrange(1,6),random.randrange(5,11))
c6=Car("Fiat", "Ascent", 2017, "Black")
c6.start()
c6.accelerate(random.randrange(1,6),random.randrange(5,11))
c7=Car("Honda", "Zen", 2011, "Blue")
c7.start()
c7.accelerate(random.randrange(1,6),random.randrange(5,11))
c8=Car("Honda", "Classic", 2012, "Black")
c8.start()
c8.accelerate(random.randrange(1,6),random.randrange(5,11))
c9=Car("Ferari", "Zen", 2014, "Blue")
c9.start()
c9.accelerate(random.randrange(1,6),random.randrange(5,11))
c10=Car("BMW", "Kaizen", 2004, "White")
c10.start()
c10.accelerate(random.randrange(1,6),random.randrange(5,11))
c6.deaccelerate(random.randrange(1,6),random.randrange(5,11))
c6.getcurrentposition()
c6.getcurrentspeed()
c9.deaccelerate(random.randrange(1,6),random.randrange(5,11))
c9.getcurrentposition()
c9.getcurrentspeed()
```
