**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```
f=open('abc.txt',  'r')
txt=f.read()
print(len(txt))
words=txt.split()
freq_chart={}
for word in words:
  curr_count= freq_chart.setdefault(word,0)
  freq_chart[word]= curr_count+1
aa=sorted(freq_chart.items(), key=lambda x:x[1] ,reverse=True)
print(aa[:100])
```

**Lab 4**

* Implement accelerate and deaccelerate methods for the car class assume of fix rate of acceleration and deaccelerate make sure car doesn’t move backward when it deaccelerating.
* Simulate a race of 10 cars. All car starts at the same position and accelerates at a random rate between 1 and 5 m/$s^2$ for a period of between 5 to 10 seconds.


```
class Car(object):
    # Constructor, takes arguments and constructs the class
    def __init__(self, year, make):
        self.__year = year    # car's year
        self.__make = make    # make of the car
        self.__speed = 0      # current speed

    def accelerate(self):
        """ Accelerate car. """
        self.__speed += 5

    def brake(self):
        """ Decelerate car. """
        self.__speed -= 5

    def get_speed(self):
        """ Get current speed of car. """
        return self.__speed

def main():
    # create a Car object
    car = Car("1990 Laser", "Plymouth")
    # calls the accelerate method
    print ('Accelerating')

    for i in range(5):
        car.accelerate()
        print (car.get_speed())

    # calls the brake method five times.
    print ("Decelerating . . .")

    for i in range(8):
        car.brake()
        print(car.get_speed())
        if car.get_speed()==0:
          break

if __name__ == "__main__":
    main()


        
```
