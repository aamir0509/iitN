**Lab 3**

* Write a program that read a text file and prints the top 100 most used words
* Hints
  * To read a file:
    * f = open(‘abc.txt’)
    * text = f.read()
  * To split text into words:
    * words = text.split()
  * To sort a list:
    * sorted(list, key=func, reverse=True)

```
f=open('abc.txt','r')
text=f.read()
words=text.split()
freq_chart = { }
for word in words:
  curr_count=freq_chart.setdefault(word,0)
  freq_chart[word]=curr_count+1
most_used=sorted(freq_chart.items(),key=lambda x:x[1],reverse=True)
print(most_used[:100])
```

* Implement accelerate and deaccelerate methods for the car class assume of fix rate of acceleration and deaccelerate make sure car doesn’t move backward when it deaccelerating.
* Simulate a race of 10 cars. All car starts at the same position and accelerates at a random rate between 1 and 5 m/$s^2$ for a period of between 5 to 10 seconds.


```
class Car:
      def __init__(self,year,make,color,speed):
        self.year = year
        self.make = make
        self.color = color
        self.speed = 0
        self.status = "stopped"
    
      def accelerate(self):
        self.speed +=5
        if (self.speed == 0) :
          self.status = "stopped"
        else:
          self.status = "Running"
 
      def deaccelerate(self):
        self.speed -=5
        if (self.speed == 0) :
          self.status = "stopped"
        else:
          self.status = "Running"
        
      def get_speed(self):
        return self.speed
   
    mycar = Car(2009,"Maruti","White",0 )
    
    print('The current status & speed is: ', mycar.status , mycar.get_speed())
    mycar.accelerate()
    print('The current status & speed is: ', mycar.status ,mycar.get_speed())
    mycar.accelerate()
    print('The current status & speed is: ', mycar.status ,mycar.get_speed())
    mycar.accelerate()
    print('The current status & speed is: ', mycar.status ,mycar.get_speed())
   
    mycar.deaccelerate()
    print('The current status & speed after deaccelerate is: ',mycar.status , mycar.get_speed())
    mycar.deaccelerate()
    print('The current status & speed after deaccelerate is: ',mycar.status , mycar.get_speed())
    mycar.deaccelerate()
    print('The current status & speed after deaccelerate is: ',mycar.status , mycar.get_speed())
```

Do object modeling for an application that manages a users personal library

```
class music_differentiator:
  def _init_(self, *music_file):
    self.music_file = music_file
 
  def mp(self):
    mp3 = []
    mp4 = []
    for x in self.music_file:
       if x._contains_('.mp3') is True: 
          mp3.append(x)
       else:
          mp4.append(x)
    print('Mp3 music file', mp3)
    print('MP4 music file', mp4)
 
ins = music_differentiator('mainepyaarkiya.mp3','kalikaliankhee.mp3','mainehoondon.mp4','hathighorapalki.mp3')
ins.mp()
```
