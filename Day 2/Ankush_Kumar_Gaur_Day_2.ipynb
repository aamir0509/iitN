**Lab 6**

Given 100 random points in 2D space and find the top 10 closest points to the origin.

```
# Given 100 random points in 2D space and find the top 10 closest points to the origin.
import numpy as np 
z= np.random.rand(100,2)
np.flip(z[np.argsort(np.sqrt(np.sum(z*z,axis=1)))],axis=0)[0:10]
```

**Lab 7**

Find the value of pi using Monte Carlo simulation.

**Hint:-** 

Consider a unit circle which is enclosed in a square. randomly genrate uniformily distributed points inside the square. The fraction of points falling inside the circle is pi/4. 

```
# Lab 7.Find the value of pi using Monte Carlo simulation.
from pylab import plot,show,axis
from numpy import random,sqrt,pi

# scattering n points over the unit square
n = 1000000
p = random.rand(n,2)

# counting the points inside the unit circle
idx = sqrt(p[:,0]**2+p[:,1]**2) < 1

# estimation of pi
print ('%0.16f' % (sum(idx).astype('double')/n*4),'result')
print ('%0.16f' % pi,'real pi')
```

```
# 2.How to get the block-sum of a 16x16 matrix (block size is 4x4)?
Z = np.ones((16,16))
k = 4
S = np.add.reduceat(np.add.reduceat(Z, np.arange(0, Z.shape[0], k), axis=0),
                                       np.arange(0, Z.shape[1], k), axis=1)
print(S)
```
