```
import pandas as pd
users=pd.read_csv("https://bit.ly/2UUomN0",delimiter='\t')
#m=pd.to_numeric(users['item_price'].str.replace('$',''))
users['item_price']=users['item_price'].map(lambda x:float(x.strip()[1:]))
users
users = users[['quantity','item_price']][users.item_price==users['item_price'].max()]
users
```

```
#veggie salad Bowl
import pandas as pd
users=pd.read_csv("https://bit.ly/2UUomN0",delimiter='\t')
users
c=users[users['item_name']=='Veggie Salad Bowl']
print(c.quantity.sum())
c=users[(users['item_name']=='Canned Soda') & (users['quantity']>1)]
print(c['quantity'].count())

```

**Lab 4**

DataSet: https://bit.ly/2CudMVh
1. Assign it to a variable called drinks.
2. Which continent drinks more beer on average?
3. For each continent print the statistics for wine consumption.
4. Display the mean alcohol consumption per continent for every column.
5. Display the median alcohol consumption per continent for every column.
6. Display the mean, min and max values for spirit consumption.

```
import pandas as pd
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv",sep=',' )
drinks.head(60)
more_avg_beer=drinks.groupby('continent').beer_servings.mean()
more_avg_beer.head(1)
drinks.groupby('continent').wine_servings.mean()
drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
drinks.groupby('continent').total_litres_of_pure_alcohol.median()
drinks.spirit_servings.mean()
drinks.spirit_servings.min(axis=0)
drinks.spirit_servings.max(axis=0)

```
