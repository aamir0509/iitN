**Lab 2**

Dataset: https://bit.ly/2UUomN0
1. What is the price of each item?
2. What was the quantity of the most expensive item ordered?
3. How many times were a Veggie Salad Bowl ordered?
4. How many times people ordered more than one Canned Soda?



```
#1.What is the price of each item?
import pandas as pd 
import numpy as np
load=pd.read_csv('https://bit.ly/2UUomN0',delimiter ='\t')
print(load[['item_name','item_price']])
```

```
#2.What was the quantity of the most expensive item ordered?
import pandas as pd 
import numpy as np
load=pd.read_csv('https://bit.ly/2UUomN0',delimiter ='\t')
print(load[['item_name','item_price']])
load['item_price']=load['item_price'].map(lambda x: float(x.strip()[1:]))
mostexp=load.sort_values(by='item_price',ascending=False)
print(mostexp)
print('\n\n\nMost expensive item Quantity')
print(mostexp[['quantity','item_name']])
```

```
# 3.How many times were a Veggie Salad Bowl ordered?
import pandas as pd 
import numpy as np
load=pd.read_csv('https://bit.ly/2UUomN0',delimiter ='\t')
print(load[['item_name','item_price']])
load['item_price']=load['item_price'].map(lambda x: float(x.strip()[1:]))
mostexp=load.sort_values(by='item_price',ascending=False)
print(mostexp)
print('\n\n\nMost expensive item Quantity')
print(mostexp[['quantity','item_name']])
vsb=list(load.item_name)

print("\n \n \n Veggie Salad Bowl ordered Following times") 
print(vsb.count("Veggie Salad Bowl")) 




```

```
# 4.How many times people ordered more than one Canned Soda?
import pandas as pd 
import numpy as np
load=pd.read_csv('https://bit.ly/2UUomN0',delimiter ='\t')
print(load[['item_name','item_price']])
load['item_price']=load['item_price'].map(lambda x: float(x.strip()[1:]))
mostexp=load.sort_values(by='item_price',ascending=False)
print(mostexp)
print('\n\n\nMost expensive item Quantity')
print(mostexp[['quantity','item_name']])
vsb=list(load.item_name)

print("\n \n \n Veggie Salad Bowl ordered Following times") 
print(vsb.count("Veggie Salad Bowl"))
soda1 = load[(load['item_name'] == 'Canned Soda') & (load['quantity'] > 1)]
print(soda1)
print("\n\n\n\n\nPeople ordered more than one Canned Soda")

print(soda1.shape[0])
```

```
# 2.Is there any missing value in the dataframe?
import numpy as np 
import pandas as pd
load=pd.read_csv('https://bit.ly/1N5br3h',delimiter =',',names=['1st','2nd','3rd','4th','class'])
print(load)
load.isnull().any()
```

```
# 3.delete the column class
import numpy as np 
import pandas as pd
load=pd.read_csv('https://bit.ly/1N5br3h',delimiter =',',names=['1st','2nd','3rd','4th','class'])
load.drop(['class'], axis=1)
```

```
# 4. set the values of the rows 10 to 29 of the column 'petal_length' to NaN
import numpy as np 
import pandas as pd
load=pd.read_csv('https://bit.ly/1N5br3h',delimiter =',',names=['sepal_length','sepal_width','petal_length','petal_width','species'])
load.iloc[10:30, -2]=np.NaN
load.head(40)

```

```
# 5.Delete the rows that have NaN
import numpy as np 
import pandas as pd
load=pd.read_csv('https://bit.ly/1N5br3h',delimiter =',',names=['sepal_length','sepal_width','petal_length','petal_width','class'])
load.iloc[10:30, -2]=np.NaN
load.dropna(axis=0,inplace=True)
load.head(n=40)
```

**Lab 4**

DataSet: https://bit.ly/2CudMVh
1. Assign it to a variable called drinks.
2. Which continent drinks more beer on average?
3. For each continent print the statistics for wine consumption.
4. Display the mean alcohol consumption per continent for every column.
5. Display the median alcohol consumption per continent for every column.
6. Display the mean, min and max values for spirit consumption.

```
# 1. Assign it to a variable called drinks.
import pandas as pd
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
print(drinks)
```

```
# 2.	Which continent drinks more beer on average?
import pandas as pd
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
drinks[['beer_servings','continent']].continent.value_counts();
more_avg_beer=drinks.groupby('continent').beer_servings.mean()
more_avg_beer.head(1)
```

```
# 3. For each continent print the statistics for wine consumption.
import pandas as pd
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
drk=drinks.dropna(axis=0)
print(drk.groupby('continent').wine_servings.describe())
```

```
#4. Display the mean alcohol consumption per continent for every column.
import pandas as pd
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
drinks.groupby('continent').mean()
```

```
#5. Display the median alcohol consumption per continent for every column
import pandas as pd
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
drinks.groupby('continent').total_litres_of_pure_alcohol.median()

```

```
#6. Display the mean, min and max values for spirit consumption.
import pandas as pd
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
print("mean for spirit consumption")
print(drinks.spirit_servings.mean())
print("min value for spirit consumption")
print(drinks.spirit_servings.max())
print("max value for spirit consumption")
print(drinks.spirit_servings.min())
```
