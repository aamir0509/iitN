**Lab 4**

DataSet: https://bit.ly/2CudMVh
1. Assign it to a variable called drinks.
2. Which continent drinks more beer on average?
3. For each continent print the statistics for wine consumption.
4. Display the mean alcohol consumption per continent for every column.
5. Display the median alcohol consumption per continent for every column.
6. Display the mean, min and max values for spirit consumption.

```
#Assign it to a variable called drinks.
import pandas as pd
drinks=pd.read_csv("https://bit.ly/2CudMVh")
drinks
```

```
#Which continent drinks more beer on average?
drinks[['beer_servings','continent']].continent.value_counts()
```

```
# For each continent print the statistics for wine consumption.
drk=drinks.dropna(axis=0)
#print(drk)
print(drk.groupby('continent').wine_servings.describe())
```

```
#Display the mean alcohol consumption per continent for every column.
drinks.groupby('continent').mean()
```

```
#Display the median alcohol consumption per continent for every column.
drinks.groupby('continent').total_litres_of_pure_alcohol.median()
```

```
#Display the mean, min and max values for spirit consumption.
print(drinks.spirit_servings.mean())        
print(drinks.spirit_servings.max())          
print(drinks.spirit_servings.min())          
```
