**Lab 3**

Dataset: https://bit.ly/1N5br3h
1. Assign Columns for the dataset 
2. Is there any missing value in the dataframe?
3. delete the column class
4. set the values of the rows 10 to 29 of the column 'petal_length' to NaN
5. Delete the rows that have NaN
6. Reset the index so it begins with 0 again
7. BONUS: Create & Answer your own question


```
#Assign Columns for the dataset
users=pd.read_csv("https://bit.ly/1N5br3h", delimiter=',', names=['sepal_length','sepal_width','petal_length','petal_width','class'])
users
```

```
#missing value in the dataframe
df=users.isnull().any()
df
```

```
#delete the column class
df_drop=users.drop(columns=['class'], axis=1)
df_drop
```

```
#set the values of the rows 10 to 29 of the column 'petal_length' to NaN
import numpy as np
df_drop.iloc[10:30, -2]=np.NaN
df_drop.head(31)
```

```
#Delete the rows that have NaN
df_drop.dropna(axis=0, inplace=True)
df_drop.head(n=31)
```

```
#Reset the index so it begins with 0 again
df_drop = df_drop.reset_index()
del df_drop['index']
df_drop.head(31)
```

**Lab 4**

DataSet: https://bit.ly/2CudMVh
1. Assign it to a variable called drinks.
2. Which continent drinks more beer on average?
3. For each continent print the statistics for wine consumption.
4. Display the mean alcohol consumption per continent for every column.
5. Display the median alcohol consumption per continent for every column.
6. Display the mean, min and max values for spirit consumption.

```
#Assign it to a variable called drinks.
import pandas as pd
dks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv",sep=',' )
dks.head(40)
```

```
#continent drinks more beer on average
mba=dks.groupby('continent').beer_servings.mean()
mba.head(1)
```

```
#For each continent print the statistics for wine consumption
dks.groupby('continent').wine_servings.mean()
```

```
#mean alcohol consumption per continent for every column.
dks.groupby('continent').total_litres_of_pure_alcohol.mean()
```

```
#the median alcohol consumption per continent for every column.
dks.groupby('continent').total_litres_of_pure_alcohol.median()
```

```
#the mean, min and max values for spirit consumption.
mn=dks['spirit_servings'].mean()
print("Mean Consumption",mn)
mnm=dks['spirit_servings'].min(axis=0)
print("Minimum Consumption",mnm)
mxn=dks['spirit_servings'].max(axis=0)
print("Mean Consumption",mxn)
```
