**Lab 1**

Using the dataset at https://bit.ly/2JIORnq
1. Assign it to a variable called users and use the 'user_id' as index
2. What is the age with least occurrence?
3. What is the mean age of users?
4. What is the most frequent occupation?

```
import pandas as pd
# reading csv file  
#readfile = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user") 
users = pd.read_csv("https://bit.ly/2JIORnq",sep='|')
print(users)
print("mean age of users")
mean1 = sum(users['age'])/len(users['age'])
print(mean1)
print("age least occurence")
leastoccur = users['age'].value_counts()
print(min(leastoccur.head(-1)))
print(leastoccur.tail(1))
print("most frequent occupation")
freqoccup = users['occupation'].value_counts()
print(freqoccup.head(1))

```
