**Lab 2**

Dataset: https://bit.ly/2UUomN0
1. What is the price of each item?
2. What was the quantity of the most expensive item ordered?
3. How many times were a Veggie Salad Bowl ordered?
4. How many times people ordered more than one Canned Soda?



```
# What is the price of each item?
import pandas as pd
import numpy as np
load=pd.read_csv('https://bit.ly/2UUomN0', delimiter ='\t')
print(load[['item_name', 'item_price']])
```

```
# What was the quantity of the most expensive item ordered?
import pandas as pd
import numpy as np
user=pd.read_csv('https://bit.ly/2UUomN0', delimiter='\t')
print(user[['item_name', 'item_price']])
user['item_price']=user['item_price'].map(lambda x: float(x.strip()[1:]))
mostexp=user.sort_values(by='item_price', ascending=False)
print(mostexp)

print('\n \n \n Most Expensive Item Quantity')
print(mostexp[['quantity', 'item_name']])
```

```
# How many times were a Veggie Salad Bowl ordered?
import pandas as pd
import numpy as np
load=pd.read_csv('https://bit.ly/2UUomN0', delimiter='\t')
print(load[['item_name','item_price']])
load['item_price']=load['item_price'].map(lambda x: float(x.strip()[1:]))
mostexp=load.sort_values(by='item_price',ascending=False)
print(mostexp)
print('\n\n\nMost expensive item Quantity')
print(mostexp[['quantity','item_name']])
vsb=list(load.item_name)

print("\n\n\n Veggie Salad Bowl ordered following times")
print(vsb.count("Veggie Salad Bowl"))
```
