**Lab 2**

Dataset: https://bit.ly/2UUomN0
1. What is the price of each item?
2. What was the quantity of the most expensive item ordered?
3. How many times were a Veggie Salad Bowl ordered?
4. How many times people ordered more than one Canned Soda?



```
#What is the price of each item?
import pandas as pd 
mydata = pd.read_csv("https://bit.ly/2UUomN0",sep='\t' )
mydata[['item_name','item_price']].drop_duplicates()
```

```
#What was the quantity of the most expensive item ordered?
import pandas as pd 
import numpy as np
mydata = pd.read_csv("https://bit.ly/2UUomN0",sep='\t' )
mydata['item_price']=mydata['item_price'].map(lambda x:float(x.strip()[1:]))
print(mydata.iloc[mydata.item_price.argmax()])
```

```
#How many times were a Veggie Salad Bowl ordered?
import pandas as pd 
import numpy as np
mydata = pd.read_csv("https://bit.ly/2UUomN0",sep='\t' )
#mydata.groupby(["item_name"]).sum().sort_values("item_price", ascending=False)
mydata.groupby("item_name").sum().sort_values("item_name", ascending=False).head()
da=mydata[mydata["item_name"]=="Veggie Salad Bowl"]
print(da.quantity.sum())
```

```
#How many times people ordered more than one Canned Soda?
mydata = pd.read_csv("https://bit.ly/2UUomN0", sep='\t')
da=mydata[(mydata["item_name"]=="Canned Soda") & (mydata['quantity']>1)]
print(da)
```

```
import pandas as pd 
drinks = pd.read_csv("https://bit.ly/2CudMVh",sep=',' )
drinks
```

```
#Which continent drinks more beer on average?
more_avg_beer=drinks.groupby('continent').beer_servings.mean()
more_avg_beer.head(1)
```

```
#For each continent print the statistics for wine consumption.
drinks.groupby('continent').wine_servings.mean()
```

```
#4.	Display the mean alcohol consumption per continent for every column.
drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
```

```
#5.	Display the median alcohol consumption per continent for every column.
drinks.groupby('continent').total_litres_of_pure_alcohol.median()
```
