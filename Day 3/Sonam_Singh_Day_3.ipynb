**Lab 4**

DataSet: https://bit.ly/2CudMVh
1. Assign it to a variable called drinks.
2. Which continent drinks more beer on average?
3. For each continent print the statistics for wine consumption.
4. Display the mean alcohol consumption per continent for every column.
5. Display the median alcohol consumption per continent for every column.
6. Display the mean, min and max values for spirit consumption.

```
import pandas as pd

drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv", sep = ',')

drinks


```

```
drinks_2 = drinks.groupby('continent').aggregate({'beer_servings': 'mean'})

drinks_2
```

```
drinks_3 = drinks.groupby('continent').aggregate({'wine_servings': ['mean','median','sum','var','std']})
drinks_3
```

```
drinks_4 = drinks.groupby('continent').aggregate({'beer_servings':'mean','spirit_servings':'mean','wine_servings':'mean'})

drinks_4
```

```
drinks_5 = drinks.groupby('continent').aggregate({'beer_servings':'median','spirit_servings':'median','wine_servings':'median'})

drinks_5
```

```
drinks_6 = drinks.groupby('continent').aggregate({'spirit_servings':['mean','min','max']})

drinks_6
```
