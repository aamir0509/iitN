1. What is the price of each item?
2. What was the quantity of the most expensive item ordered?
3. How many times were a Veggie Salad Bowl ordered?
4. How many times people ordered more than one Canned Soda?



```
import pandas as pd
users=pd.read_csv('https://bit.ly/2UUomN0',delimiter='\t')
users['item_price']=users['item_price'].str.replace('$',' ')
users['item_actual_price']=users['item_price'].astype(float)/users['quantity']
users.drop_duplicates(subset='item_name', keep='first', inplace=False)
users
```

```
import pandas as pd
users=pd.read_csv('https://bit.ly/2UUomN0',delimiter='\t')
users['item_price']=users['item_price'].str.replace('$',' ')
users['item_actual_price']=users['item_price'].astype(float)/users['quantity']
result = users.sort_values('item_actual_price', ascending=False)
result.drop_duplicates(subset='item_name', keep='first',inplace=True)
result[:3]
```

DataSet: https://bit.ly/2CudMVh
1. Assign it to a variable called drinks.
2. Which continent drinks more beer on average?
3. For each continent print the statistics for wine consumption.
4. Display the mean alcohol consumption per continent for every column.
5. Display the median alcohol consumption per continent for every column.

```
import pandas as pd
drinks=pd.read_csv('https://bit.ly/2CudMVh', delimiter=',')
drinks
```

```
most_servings=drinks.groupby(by='continent').mean().sort_values('beer_servings',ascending=False)
most_servings[:1]
```

```
#For each continent print the statistics for wine consumption
import pandas as pd
drinks=pd.read_csv('https://bit.ly/2CudMVh',delimiter=',')
wine_data=drinks.groupby(by='continent').describe()
wine_data['wine_servings']
```

```
#Display the mean alcohol consumption per continent for every column.
most_servings
```

```
#Display the median alcohol consumption per continent for every column.
most_alcohol=drinks.groupby(by='continent').median().sort_values('total_litres_of_pure_alcohol',ascending=False)
most_alcohol['total_litres_of_pure_alcohol']
```

```
#Display the mean, min and max values for spirit consumption
import pandas as pd
drinks=pd.read_csv('https://bit.ly/2CudMVh',delimiter=',')
spirit_data=drinks['spirit_servings']
spirit_data.describe()[['min','max','mean']]
```
